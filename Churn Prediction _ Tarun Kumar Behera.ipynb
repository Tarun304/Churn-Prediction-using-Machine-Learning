**Churn Definition and Reasoning:**

Churn, in the context of this e-commerce platform, refers to the disengagement of users from meaningful platform interactions, primarily focusing on purchasing behavior. A robust churn definition is essential to identify users at risk of disengagement, enabling targeted retention strategies and revenue protection.

**Key Considerations:**

To define churn effectively, we considered the following factors:

*1. User Behavior Patterns:*

* Users exhibit varying levels of engagement, from frequent purchasers to sporadic visitors.
*Engagement includes non-purchase activities (e.g., viewing or adding items to the cart) and purchases.

*2. Business Objectives:*
* Retain users contributing to revenue by identifying those showing signs of disengagement.
* Ensure the churn definition aligns with historical user behavior and typical purchasing cycles.

*3. Time-Based Thresholds:*
* Churn often involves a lack of activity over a specific period.
* Historical purchase frequency and platform interaction patterns informed threshold selection.

**Proposed Churn Definition:**
A user is considered churned if they meet one or both of the following criteria:

*1. No Purchase Activity:*

* The user has not made a purchase within the last 30 days (purchase threshold).
* This captures users deviating significantly from typical purchasing behavior.

*2. Overall Inactivity:*

* The user has not performed any activity (view, cart, or purchase) within the last 90 days (activity threshold).
* This ensures that churn captures broader disengagement from the platform.

**Rationable Behind The Above Definition:**
* *Dual Metrics:* Combines purchase inactivity with overall platform engagement for comprehensive coverage of disengagement.
* *Threshold Selection:*
30-day purchase threshold reflects typical e-commerce purchasing cycles.
90-day activity threshold captures sporadic yet valuable users.
* *Flexibility:* Thresholds can be tailored to user behavior or business needs.
* *Business Implications:* Identifies at-risk users for precise re-engagement strategies, balancing scalability and behavioral relevance.



```python
# Step 1: Aggregate user activity
user_activity = df.groupby('user_id').agg(
    last_event_time=('event_time', 'max'),
    last_purchase_time=('event_time', lambda x: x[df.loc[x.index, 'event_type'] == 'purchase'].max() if any(df.loc[x.index, 'event_type'] == 'purchase') else pd.NaT)
).reset_index()

# Step 2: Calculate recency for both activity and purchase
current_date = df['event_time'].max()
user_activity['days_since_last_activity'] = (current_date - user_activity['last_event_time']).dt.days
user_activity['days_since_last_purchase'] = (current_date - user_activity['last_purchase_time']).dt.days

# Step 3: Define churn thresholds
purchase_threshold = 30  # Days since last purchase
activity_threshold = 90  # Days since last activity

# Step 4: Create churn label
user_activity['churn'] = ((user_activity['days_since_last_purchase'] > purchase_threshold) |
                          (user_activity['days_since_last_activity'] > activity_threshold)).astype(int)

# Step 5: Merge churn label back to the main dataset
df = df.merge(user_activity[['user_id', 'churn']], on='user_id', how='left')

# Step 6: Display churn distribution
print("Churn Distribution:")
print(user_activity['churn'].value_counts())
```

**Feature Engineering:**

**Steps taken for feature engineering:**
1. Aggregate user-level data: Total events, views, purchases, recency, and derived ratios.
2. Calculate top product/brand preferences by views and purchases.
3. Aggregate session-level data and calculated session-based metrics.
4. Merge all user-level and session-level metrics into the original DataFrame.

```python
import pandas as pd
import numpy as np

# Aggregate user-level data
user_features = df.groupby('user_id').agg(
    total_events=('user_id', 'size'),  # Total number of events for each user
    total_views=('event_type_view', 'sum'),  # Total 'view' events from one-hot encoded column
    total_purchases=('event_type_purchase', 'sum'),  # Total 'purchase' events from one-hot encoded column
    avg_session_duration=('event_time', lambda x: (x.max() - x.min()).total_seconds() / len(x)),  # Avg session duration
    last_event_time=('event_time', 'max'),  # Recency calculation (last activity time)
).reset_index()

# Calculate derived ratios for user-level features
user_features['purchase_to_view_ratio'] = user_features['total_purchases'] / (user_features['total_views'] + 1e-5)
user_features['recency_days'] = (df['event_time'].max() - user_features['last_event_time']).dt.days  # Days since last activity


```

```python
# Calculate top viewed category
top_viewed = df.groupby('user_id')['category_code'].agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan).reset_index()
top_viewed.rename(columns={'category_code': 'top_viewed_category'}, inplace=True)

# Filter for purchase events and calculate top purchased category
purchased_df = df[df['event_type_purchase']]
top_purchased = purchased_df.groupby('user_id')['category_code'].agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan).reset_index()
top_purchased.rename(columns={'category_code': 'top_purchased_category'}, inplace=True)

# Merge the results into a single DataFrame
user_product_preferences = top_viewed.merge(top_purchased, on='user_id', how='left')

# Merge product preferences into user_features
user_features = user_features.merge(user_product_preferences, on='user_id', how='left')
```

```python
# Step 1: Calculate session-level features


session_features = df.groupby('user_session').agg(
    session_length=('event_time', lambda x: (x.max() - x.min()).total_seconds()),  # Session duration
    session_events=('event_time', 'size')  # Number of events in a session
).reset_index()

# Step 2: Add session-level metrics back to the main DataFrame for easy aggregation
df = df.merge(session_features, on='user_session', how='left')

# Step 3: Aggregate user-level session metrics
user_session_metrics = df.groupby('user_id').agg(
    total_sessions=('user_session', 'nunique'),  # Total number of sessions
    avg_session_length=('session_length', 'mean'),  # Average session duration
    bounce_rate=('session_events', lambda x: np.mean(x <= 1))  # Bounce rate: proportion of single-event sessions
).reset_index()

# Step 4: Merge session metrics into user_features
user_features = user_features.merge(user_session_metrics, on='user_id', how='left')

# Step 5: Merge user-level metrics back into the original DataFrame
df = df.merge(user_features, on='user_id', how='left')

```

 Features Created:

   - `total_events`, `total_views`, `total_purchases`, `avg_session_duration`, `last_event_time`, `recency_days`, and `purchase_to_view_ratio` highlight user activity and intent.
   - Top viewed and purchased categories identify user preferences for personalized interventions.

2. **Session-Level Metrics:**
   - `session_length`, `session_events`, `total_sessions`, `avg_session_length`, and `bounce_rate` measure engagement depth and interaction quality.

3. **Temporal Features:**
   - `event_day`, `event_hour`, and `event_month` capture seasonal and hourly trends.

**Model Building and Evaluation:**

We have chosen two models : **XgBoost** and **LightGBM**. Due to lack of resources , it is not possible to fine tune these models using Optuna or GridSearchCV. Hence, we will train these directly with the manually defined hyperparameters.

Model : XgBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
import pandas as pd

# Train XGBoost model
xgb = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    n_estimators=200,
    learning_rate=0.03,  # Reduced learning rate
    max_depth=3,         # Reduced max depth
    min_child_weight=5,  # Require minimum samples per leaf
    subsample=0.8,       # Use 80% of data for each tree
    colsample_bytree=0.8, # Use 80% of features for each tree
    reg_alpha=0.5,        # Increased L1 regularization
    reg_lambda=1.5        # Increased L2 regularization
)
xgb.fit(X_train_selected, y_train_smote)

# Make predictions
y_pred_xgb = xgb.predict(X_test_selected)

# Evaluate XGBoost metrics
metrics = {
    'Accuracy': accuracy_score(y_test, y_pred_xgb),
    'Precision': precision_score(y_test, y_pred_xgb),
    'Recall': recall_score(y_test, y_pred_xgb),
    'F1 Score': f1_score(y_test, y_pred_xgb),
    'Confusion Matrix': confusion_matrix(y_test, y_pred_xgb),
    'AUROC Score': roc_auc_score(y_test, xgb.predict_proba(X_test_selected)[:, 1])
}

# Convert metrics to DataFrame for better visualization
metrics_df = pd.DataFrame([metrics], index=['XGBoost'])
print("XGBoost Model Performance Metrics:")
metrics_df
```

Perform K-fold cross validation:
